```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
# Histogram when we have large dataset and we want to categorize them in bins
```

```python
data = pd.read_csv('data2.csv')
data
```

```python
#basic histogram
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.hist(ages)
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()
```

```python
#basic histogram with bins and edgecolors
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.hist(ages, bins=5, edgecolor='black')
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.tight_layout()
```

```python
#basic histogram with bins as set of values
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10,20,30,40,50,60]
plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.tight_layout()
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [30,40,50,60]
plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.tight_layout()
```
