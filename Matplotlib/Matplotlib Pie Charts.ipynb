```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
# basic pie chart
slices1 = [10,20,30,40]
plt.pie(slices1)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# basic pie chart with labels
slices1 = [10,20,30,40]
labels1 = ['Ten', 'Twenty', 'Thirty', 'Forty']
plt.pie(slices1, labels = labels1)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# basic pie chart with labels with colors
slices1 = [10,20,30,40]
labels1 = ['Ten', 'Twenty', 'Thirty', 'Forty']
colors = ['red','blue','green','yellow']
plt.pie(slices1, labels = labels1, colors=colors)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# basic pie chart with labels with colors and edges
slices1 = [10,20,30,40]
labels1 = ['Ten', 'Twenty', 'Thirty', 'Forty']
colors = ['red','blue','green','yellow']
plt.pie(slices1, labels = labels1, colors=colors ,wedgeprops={'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
plt.pie(slices, labels = labels, wedgeprops={'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# Pie charts are good for data with 5 or less items
plt.pie(slices[:5], labels = labels[:5], wedgeprops={'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
#emphasize a particular slice
explode = [0,0,0,0.1,0]
plt.pie(slices[:5], labels = labels[:5], wedgeprops={'edgecolor': 'black'}, explode=explode)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
#emphasize a particular slice
explode = [0,0,0,0.2,0]
plt.pie(slices[:5], labels = labels[:5], wedgeprops={'edgecolor': 'black'}, explode=explode)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
# Rotate and add a shadow
explode = [0,0,0,0.2,0]
plt.pie(slices[:5], labels = labels[:5], wedgeprops={'edgecolor': 'black'}, explode=explode,shadow=True,
        startangle=90)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
#add %age to your pie chart
explode = [0,0,0,0.2,0]
plt.pie(slices[:5], labels = labels[:5], wedgeprops={'edgecolor': 'black'}, explode=explode,shadow=True,
        startangle=90,autopct='%1.1f%%')
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```
