```python
import pandas as pd
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
# Read a Summary CSV file into a dataframe
dfSummary = pd.read_csv('data/dataset_summary.csv')
dfSummary
```

```python
# Read a Summary CSV file into a dataframe
dfStockPrices = pd.read_csv('data/stocks_latest/stock_prices_latest.csv')
dfStockPrices
```

```python
# Read a Dividend CSV file into a dataframe
dfDividend = pd.read_csv('data/stocks_latest/dividends_latest.csv')
dfDividend
```

```python
# Read a Dividend CSV file into a dataframe
dfEarning = pd.read_csv('data/stocks_latest/earnings_latest.csv')
dfEarning
```

```python
filterSummary = dfSummary['symbol'] == 'APPL'
dfSummary = dfSummary[filterSummary]
dfSummary
```

```python
filtertSockPrices = ((dfStockPrices['date']> '2000-01-01') & (dfStockPrices['symbol'] == 'APPL'))
dfStockPrices = dfStockPrices[filtertSockPrices].sort_values('date', ascending = False)
dfStockPrices
```

```python
filtertDividend = ((dfDividend['date']> '2000-01-01') & (dfDividend['symbol'] == 'APPL'))
dfDividend = dfDividend[filtertDividend].sort_values('date', ascending = False)
dfDividend
```

```python
filtertEarning = ((dfEarning['date']> '2000-01-01') & (dfEarning['symbol'] == 'APPL'))
dfEarning = dfEarning[filtertEarning].sort_values('date', ascending = False)
dfEarning
```

```python
s1 = pd.merge(dfEarning, dfStockPrices, how='left', on=['date'])
s1
```

```python
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')
len(s1.index)
```

```python
dfNew = pd.DataFrame([s1['date'], s1['eps']]).transpose()
dfNew
```

```python
plt.plot_date(s1['date'], s1['eps'], linestyle='solid')
```
