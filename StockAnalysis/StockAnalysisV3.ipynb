```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
# Read a Dividend CSV file into a dataframe
dfEarning = pd.read_csv('data/stocks_latest/earnings_latest.csv')
dfEarning
```

```python
filtertEarning = ((dfEarning['date']> '2000-01-01') & (dfEarning['symbol'] == 'MSFT'))
dfEarning = dfEarning[filtertEarning].sort_values('date', ascending = False)
dfEarning['eps'] = dfEarning['eps'].fillna(0)
dfEarning['eps_est'] = dfEarning['eps_est'].fillna(0)
dfEarning['eps'] = dfEarning['eps'].astype(float)
dfEarning['eps_est'] = dfEarning['eps_est'].astype(float)
dfEarning
```

```python
#dfEarning1 = pd.DataFrame([dfEarning['date'],dfEarning['date'].str.slice(0, 4), dfEarning['eps'], dfEarning['eps_est']]).transpose()
dfEarning1 = pd.DataFrame([dfEarning['date'].str.slice(0, 4), dfEarning['eps'], dfEarning['eps_est']]).transpose()
dfEarning1['eps'] = dfEarning1['eps'].astype(float)
dfEarning1['eps_est'] = dfEarning1['eps_est'].astype(float)
dfEarning1
```
