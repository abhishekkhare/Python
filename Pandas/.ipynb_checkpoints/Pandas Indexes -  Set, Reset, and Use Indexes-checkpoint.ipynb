```python
# Read a CSV file into a dataframe and modify the index column
df2 = pd.read_csv('data/friends.csv', index_col = 'Email')
df2
```

```python
df2.sort_index() # sort by index, no inplace
```

```python
df2.sort_index(ascending = False) # no inplace
```

```python
df2
```

```python
df2.sort_index(ascending = False, inplace = True) # with inplace
```

```python
df = pd.read_csv('data/olympics.csv', index_col = 0, skiprows=1)
df.head()
```

```python
df.columns
```

```python
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 

df.head()
```
